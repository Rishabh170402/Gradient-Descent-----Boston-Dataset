```python
def gdStep(data, alpha, m):
    X = data[:, :len(data[0]) - 1]
    Y = data[:, -1]
    m_slope = np.zeros(len(X[0]))
    N = len(data)
    for i in range(N):
        for j in range(len(X[0])):
            m_slope[j] += (-2/N) * (Y[i] - (m * X[i]).sum()) * X[i][j]
    new_m = m - (alpha * m_slope)
    return new_m
```

```python
def gd(iteration, alpha, data):
    X = data[:, :len(data[0]) - 1]
    Y = data[:, -1]
    m = np.zeros(len(X[0]))
    for i in range(iteration):
        m = gdStep(data, alpha, m)
        print("cost: " , cost(data, m))
    return m
```

```python
def cost(data, m):
    Tcost = 0
    N = len(data)
    X = data[:, :len(data[0]) - 1]
    Y = data[:, -1]
    for i in range(N):
        Tcost += (1/N) * ((Y[i] - ((m * X[i]).sum())) ** 2)
    return Tcost
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
def run():
    data = np.loadtxt("training.csv", delimiter = ",")
    X = data[:, :len(data[0]) - 1]
    Y = data[:, -1].reshape(len(X), 1)
    scaler.fit(X)
    X = scaler.transform(X)
    data = np.append(X, Y, axis = 1)
    
    c = np.ones(len(X)).reshape(len(X), 1)
    data = np.append(c, data, axis = 1)
    
    
    iteration = 300
    alpha = 0.1
    m = gd(iteration, alpha, data)
    print("m = ", m)
    return m
```
